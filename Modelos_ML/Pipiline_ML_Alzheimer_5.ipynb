```python
import mne
import os
import glob
import pandas as pd
from collections import Counter
import warnings

# Silenciar avisos do MNE
warnings.filterwarnings('ignore')

# 1. Configuração de Caminhos e Arquivos
caminho_raiz = r"C:\Users\tiago\Downloads\Dataset_EEG_Alzheimer\dataset_eeg_alzheimer"
arq_demografico = os.path.join(caminho_raiz, "demographics_ad_eeg_data.csv")
arq_cognitivo = os.path.join(caminho_raiz, "cognition_ad_eeg_data.csv")

# 2. Carregamento de Metadados Externos (Demografia e Cognição)
try:
    df_demo = pd.read_csv(arq_demografico)
    df_cog = pd.read_csv(arq_cognitivo)
    
    # Padronização de IDs para o Merge (removendo prefixos se existirem)
    def clean_id(df):
        col_id = [c for c in df.columns if 'id' in c.lower()][0]
        df[col_id] = df[col_id].astype(str).str.replace('sub-', '')
        return df, col_id

    df_demo, col_id_demo = clean_id(df_demo)
    df_cog, col_id_cog = clean_id(df_cog)
    
    print(f"✅ Dados externos carregados: {len(df_demo)} sujeitos identificados.")
except Exception as e:
    print(f"❌ Erro ao carregar metadados CSV: {e}")

# 3. Processamento dos Sinais EEG (.set)
padrao_busca = os.path.join(caminho_raiz, "**", "*.set")
arquivos_eeg = glob.glob(padrao_busca, recursive=True)
lista_tecnica = []

print(f"🔄 Analisando sinais de {len(arquivos_eeg)} arquivos...")

for caminho_arq in arquivos_eeg:
    nome_arq = os.path.basename(caminho_arq)
    origem = "AR" if r"\AR" in caminho_arq else "CL"
    
    try:
        raw = mne.io.read_raw_eeglab(caminho_arq, preload=False, verbose=False)
        info = raw.info
        
        # Identificação do ID para o merge
        id_limpo = nome_arq.split('_')[1].replace('sub-', '') if '_' in nome_arq else nome_arq.replace('sub-', '')
        
        # --- Lógica para eliminar redundância de canais ---
        total_ch = info['nchan']
        tipos = Counter(raw.get_channel_types())
        qtd_eeg = tipos.get('eeg', 0)
        
        # Se todos forem EEG, coloca apenas o número. Se houver outros (EOG/ECG), ele detalha.
        if total_ch == qtd_eeg:
            info_canais = f"{total_ch} (Todos EEG)"
        else:
            outros = total_ch - qtd_eeg
            info_canais = f"{total_ch} ({qtd_eeg} EEG / {outros} Outros)"

        dados_tecnicos = {
            "ID_Merge": id_limpo,
            "Arquivo": nome_arq,
            "Pais": origem,
            "Duração (s)": round(raw.times[-1], 2),
            "Freq_Amostragem (Hz)": info['sfreq'],
            "Canais_Info": info_canais,  # Coluna consolidada
            "Filtro_HP": info.get('highpass'),
            "Filtro_LP": info.get('lowpass'),
            "Bad_Channels": len(info['bads'])
        }
        lista_tecnica.append(dados_tecnicos)
    except Exception as e:
        print(f"⚠️ Erro no arquivo {nome_arq}: {e}")

# 4. Fusão de Dados (Merge: Técnica + Demografia + Cognição)
df_tecnico = pd.DataFrame(lista_tecnica)

# Merge com Demografia
df_final = pd.merge(df_tecnico, df_demo, left_on="ID_Merge", right_on=col_id_demo, how="left")

# Merge com Cognição (MMSE e outros testes)
df_final = pd.merge(df_final, df_cog, left_on="ID_Merge", right_on=col_id_cog, how="left")

# Limpeza de colunas de ID duplicadas
cols_remover = [col_id_demo, col_id_cog, "ID_Merge"]
df_final = df_final.drop(columns=[c for c in cols_remover if c in df_final.columns])

# 5. Organização e Exibição Final
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

caminho_salvamento = os.path.join(caminho_raiz, "metadados_propios_eeg_alzheimer.csv")
df_final.to_csv(caminho_salvamento, index=False)

print("\n" + "="*80)
print(f"SUCESSO! Tabela consolidada salva em: {caminho_salvamento}")
print("="*80 + "\n")

display(df_final)
```

```python
# ============================================================
# CÉLULA 1 – IMPORTAÇÕES E CONFIGURAÇÕES INICIAIS
# ============================================================
import re
import traceback
from pathlib import Path
import warnings
import mne
import numpy as np
import pandas as pd
```

```python
# ============================================================
# CÉLULA 2 – DEFINIÇÃO DE CONSTANTES
# ============================================================
PASTA_AD = r"C:\Users\tiago\Downloads\Dataset_EEG_Alzheimer\dataset_eeg_alzheimer"
PASTA_HC = r"C:\Users\tiago\Downloads\Dataset_EEG_Alzheimer\dataset_eeg_hc"
OUT_CSV_FULL = "eeg_features_brainlat_FULL_normalizado.csv"
OUT_FAILS = "eeg_features_brainlat_falhas_normalizado.csv"

# Bandas de frequência de interesse (em Hz)
BANDS = {
    "Delta": (0.5, 4),
    "Theta": (4, 8),
    "Alpha": (8, 13),
    "Beta":  (13, 30),
    "Gamma": (30, 45),
}
FMIN = 0.5
FMAX = 45
```

```python
# ============================================================
# CÉLULA 3 – FUNÇÕES AUXILIARES PARA IDENTIFICAÇÃO E CARREGAMENTO
# ============================================================
def extrair_subject_id(arq_path: Path) -> str:
    """Extrai o identificador do sujeito (ex: sub-001) do nome do arquivo."""
    m = re.search(r"(sub-\d+)", arq_path.stem, flags=re.IGNORECASE)
    if m:
        return m.group(1)
    m = re.search(r"sub[-_]?(\d+)", arq_path.stem, flags=re.IGNORECASE)
    if m:
        return f"sub-{m.group(1)}"
    return arq_path.stem

def inferir_pais(arq_path: Path) -> str:
    """Infere o país de origem a partir do caminho do arquivo (AR = Argentina, CL = Chile)."""
    parts = {p.upper() for p in arq_path.parts}
    if "AR" in parts:
        return "Argentina"
    if "CL" in parts:
        return "Chile"
    return "Outro"

def carregar_raw_seguro(arq_path: Path):
    """
    Carrega um arquivo .set do EEGLAB de forma robusta.
    Se o erro for relacionado ao arquivo .fdt ausente, tenta corrigir com cópia temporária.
    """
    try:
        return mne.io.read_raw_eeglab(str(arq_path), preload=True, verbose=False)
    except Exception as e:
        msg = str(e).lower()
        if ("fdt" in msg) or ("not found" in msg):
            fdt_exato = arq_path.with_suffix(".fdt")
            if fdt_exato.exists():
                from tempfile import TemporaryDirectory
                with TemporaryDirectory() as tmp:
                    tmp_p = Path(tmp)
                    set_tmp = tmp_p / arq_path.name
                    fdt_tmp = tmp_p / fdt_exato.name
                    set_tmp.write_bytes(arq_path.read_bytes())
                    fdt_tmp.write_bytes(fdt_exato.read_bytes())
                    return mne.io.read_raw_eeglab(str(set_tmp), preload=True, verbose=False)
        raise e
```

```python
# ============================================================
# CÉLULA 4 – EXTRAÇÃO DE MÉTRICAS ESPECTRAIS
# ============================================================
def extrair_metricas_eeg(data, sfreq):
    """
    Calcula features espectrais para uma época EEG.
    
    Parâmetros:
        data: array 2D (canais, amostras) já normalizado (RMS global ~1)
        sfreq: frequência de amostragem do sinal
    
    Retorna:
        Dicionário com potências relativas, razões e entropia espectral.
    """
    from mne.time_frequency import psd_array_welch
    data = np.asarray(data, dtype=float)
    if data.ndim != 2:
        raise ValueError(f"Esperado array 2D, recebido shape={data.shape}")
    
    # Estimativa do espectro de potência via Welch
    n_fft = min(512, data.shape[-1])
    psds, freqs = psd_array_welch(
        data, sfreq=sfreq, fmin=FMIN, fmax=FMAX,
        n_fft=n_fft, verbose=False
    )
    psd_media = np.mean(psds, axis=0)  # média sobre canais
    
    # Função de integração (trapézio)
    metodo_area = np.trapezoid if hasattr(np, "trapezoid") else np.trapz

    # Potência absoluta em cada banda
    pots = {}
    for banda, (fmin, fmax) in BANDS.items():
        idx = (freqs >= fmin) & (freqs < fmax)
        if np.sum(idx) >= 2:
            pots[banda] = float(metodo_area(psd_media[idx], freqs[idx]))
        else:
            pots[banda] = 0.0
            if banda == "Delta":
                print(f"  ⚠ Atenção: poucos pontos na banda Delta (freqs disponíveis: {freqs})")

    total = sum(pots.values()) + 1e-12  # evita divisão por zero
    
    # Features relativas
    f = {
        "Rel_Delta_mean": pots["Delta"] / total,
        "Rel_Theta_mean": pots["Theta"] / total,
        "Rel_Alpha_mean": pots["Alpha"] / total,
        "Rel_Beta_mean":  pots["Beta"] / total,
        "Rel_Gamma_mean": pots["Gamma"] / total,
    }
    f["Razao_Theta_Alpha"] = f["Rel_Theta_mean"] / (f["Rel_Alpha_mean"] + 1e-12)
    f["Razao_Theta_Beta"]  = f["Rel_Theta_mean"] / (f["Rel_Beta_mean"] + 1e-12)

    # Entropia espectral
    psd_norm = psd_media / (np.sum(psd_media) + 1e-12)
    f["Spectral_Entropy"] = float(-np.sum(psd_norm * np.log2(psd_norm + 1e-12)))
    
    return f
```

```python
# ============================================================
# CÉLULA 5 – PROCESSAMENTO PRINCIPAL E EXTRAÇÃO DE FEATURES
# ============================================================
dados_finais = []
falhas = []

for grupo, pasta in [("AD", PASTA_AD), ("HC", PASTA_HC)]:
    arquivos = list(Path(pasta).rglob("*.set"))
    print(f"\n📂 {grupo}: {len(arquivos)} arquivos.")
    
    for arq in arquivos:
        try:
            sujeito_id = extrair_subject_id(arq)
            pais = inferir_pais(arq)
            raw = carregar_raw_seguro(arq)

            # 1. Re‑referenciação para a média
            raw.set_eeg_reference('average', projection=False)

            # 2. Filtro passa‑faixa (0.5 - 45 Hz)
            raw.filter(0.5, 45, fir_design='firwin', verbose=False)

            # 3. Selecionar apenas canais EEG
            raw.pick_types(eeg=True, eog=False, ecg=False, stim=False, emg=False, misc=False)

            # 4. Criar épocas de 4 segundos
            epochs = mne.make_fixed_length_epochs(raw, duration=4.0, preload=True, verbose=False)
            if len(epochs) == 0:
                raise RuntimeError("Nenhuma época gerada.")

            data = epochs.get_data()   # (n_epochs, n_channels, n_times)
            sfreq = float(raw.info["sfreq"])

            # 5. NORMALIZAÇÃO DE AMPLITUDE POR SUJEITO (global RMS)
            all_data = data.reshape(-1, data.shape[-1])
            global_rms = np.sqrt(np.mean(all_data**2))
            if global_rms < 1e-12:
                global_rms = 1.0
            scale = 1.0 / global_rms
            data_normalized = data * scale   # RMS global = 1

            # 6. Extrair features época a época
            for i, epoca in enumerate(data_normalized):
                feat = extrair_metricas_eeg(epoca, sfreq)
                feat.update({
                    "subject_id": sujeito_id,
                    "label": grupo,
                    "Pais": pais,
                    "epoch_id": i,
                    "source_file": arq.name,
                })
                dados_finais.append(feat)

            print(f"  ✅ {sujeito_id} ({len(data)} épocas) | RMS global = {global_rms:.6f} -> escala = {scale:.2f}")
        
        except Exception as e:
            falhas.append({
                "grupo": grupo,
                "arquivo": arq.name,
                "caminho": str(arq),
                "erro": str(e),
                "traceback": traceback.format_exc(),
            })
            print(f"  ❌ {arq.name} → {e}")
```

```python
# ============================================================
# CÉLULA 8 – FUNÇÃO AUXILIAR DE NORMALIZAÇÃO (FIT NO TREINO)
# ============================================================
def aplicar_standard_scaler(X_tr, X_te, cols):
    """
    Aplica StandardScaler de forma correta para evitar vazamento de dados.
    
    Parâmetros:
        X_tr : DataFrame de treino
        X_te : DataFrame de teste
        cols : lista de colunas a serem normalizadas
    
    Retorna:
        X_tr_out, X_te_out : DataFrames com as colunas normalizadas
    """
    scaler   = StandardScaler()
    X_tr_out = X_tr.copy()
    X_te_out = X_te.copy()
    X_tr_out[cols] = scaler.fit_transform(X_tr[cols])   # Ajusta apenas no treino
    X_te_out[cols] = scaler.transform(X_te[cols])       # Aplica no teste
    return X_tr_out, X_te_out
```

```python
# ============================================================
# CÉLULA 9 – FUNÇÃO PRINCIPAL DE VALIDAÇÃO LOSO
# ============================================================
def loso_correto(X_df, y_arr, groups_arr, modelo_base,
                 feature_cols=FEATURE_COLS, verbose=True):
    """
    Leave-One-Subject-Out sem vazamento de pré-processamento.

    Parâmetros
    ----------
    X_df         : pd.DataFrame — épocas × features
    y_arr        : array-like   — rótulo binário por época (0=HC, 1=AD)
    groups_arr   : array-like   — subject_id por época
    modelo_base  : estimador sklearn — clonado a cada fold
    feature_cols : lista de features usadas no treino/predição
    verbose      : bool — imprime progresso por fold

    Retorna
    -------
    pd.DataFrame com colunas [subject_id, real, proba, n_epocas]
    ou None se nenhum fold for válido.
    """

    # ── Validações iniciais ──────────────────────────────────
    X_df       = X_df.reset_index(drop=True)
    y_arr      = np.asarray(y_arr).ravel()
    groups_arr = np.asarray(groups_arr).ravel()

    if not (len(X_df) == len(y_arr) == len(groups_arr)):
        raise ValueError(
            f"Tamanhos incompatíveis: X={len(X_df)}, "
            f"y={len(y_arr)}, groups={len(groups_arr)}"
        )

    cols_ausentes = [c for c in feature_cols if c not in X_df.columns]
    if cols_ausentes:
        raise ValueError(f"Features ausentes no DataFrame: {cols_ausentes}")

    # ── Auditoria de variância antes de treinar ───────────────
    std_zero = [c for c in feature_cols if X_df[c].std() < 1e-8]
    if std_zero:
        raise ValueError(
            f"Features com variância zero detectadas (artefato): {std_zero}\n"
            f"Remova-as de FEATURE_COLS antes de continuar."
        )

    logo       = LeaveOneGroupOut()
    preds      = []
    skipped    = []
    n_subjects = len(np.unique(groups_arr))

    if verbose:
        print(f"   Iniciando {n_subjects} folds LOSO...\n")

    # ── Loop principal ───────────────────────────────────────
    for fold_i, (train_idx, test_idx) in enumerate(
            logo.split(X_df, y_arr, groups=groups_arr)):

        subj_id = groups_arr[test_idx][0]

        X_tr_raw = X_df.iloc[train_idx]
        X_te_raw = X_df.iloc[test_idx]
        y_tr     = y_arr[train_idx]
        y_te     = y_arr[test_idx]

        # Fold inválido: treino sem as duas classes
        if len(np.unique(y_tr)) < 2:
            skipped.append(subj_id)
            if verbose:
                print(f"  ⚠  Fold {fold_i+1:02d} [{subj_id}] "
                      f"ignorado — treino com classe única")
            continue

        # [1] Normalização: fit APENAS no treino
        X_tr_sc, X_te_sc = aplicar_standard_scaler(
            X_tr_raw, X_te_raw, feature_cols
        )

        # [2] Modelo clonado (estado virgem a cada fold)
        modelo = clone(modelo_base)
        modelo.fit(X_tr_sc[feature_cols], y_tr)

        # [3] Predição por época
        proba_epocas = modelo.predict_proba(
            X_te_sc[feature_cols]
        )[:, 1]

        # [4] Agregação por sujeito (média das épocas)
        proba_sujeito = float(np.mean(proba_epocas))
        label_sujeito = int(y_te[0])

        preds.append({
            "subject_id" : subj_id,
            "real"       : label_sujeito,
            "proba"      : proba_sujeito,
            "n_epocas"   : len(proba_epocas),
        })

        if verbose:
            acerto = "✓" if int(proba_sujeito >= 0.5) == label_sujeito else "✗"
            print(f"  {acerto}  [{fold_i+1:02d}/{n_subjects}] "
                  f"{subj_id:<18} | "
                  f"real={'AD' if label_sujeito else 'HC'} | "
                  f"proba={proba_sujeito:.3f} | "
                  f"n_épocas={len(proba_epocas)}")

    # ── Resultado final ──────────────────────────────────────
    if len(preds) == 0:
        print("  ⚠  Nenhuma predição gerada — todos os folds foram ignorados.")
        return None

    if skipped and verbose:
        print(f"\n  ⚠  {len(skipped)} fold(s) ignorado(s): {skipped}")

    df_preds = pd.DataFrame(preds)

    # Calcula métricas globais e armazena nos atributos do DataFrame
    pred_bin = (df_preds["proba"] >= 0.5).astype(int)
    df_preds.attrs["auc"]  = roc_auc_score(df_preds["real"], df_preds["proba"])
    df_preds.attrs["sens"] = recall_score(df_preds["real"], pred_bin,
                                          pos_label=1, zero_division=0)
    df_preds.attrs["spec"] = recall_score(df_preds["real"], pred_bin,
                                          pos_label=0, zero_division=0)
    df_preds.attrs["f1"]   = f1_score(df_preds["real"], pred_bin,
                                      zero_division=0)

    return df_preds
```

```python
# ============================================================
# CÉLULA ÚNICA – ANÁLISE COMPLETA DE EXPLICABILIDADE E VALIDAÇÃO NEUROBIOLÓGICA
# ============================================================
# Requer: shap, sklearn, matplotlib, seaborn, pingouin (opcional para ICC)
# Instale com: pip install shap pingouin

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    roc_auc_score, roc_curve, accuracy_score, balanced_accuracy_score,
    recall_score, precision_score, f1_score, average_precision_score,
    confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.inspection import permutation_importance, PartialDependenceDisplay
from sklearn.model_selection import LeaveOneGroupOut
import shap
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ========== CONFIGURAÇÕES ==========
RANDOM_STATE = 42
N_REPEATS_PERM = 10          # repetições para permutation importance
N_BOOTSTRAP = 1000           # para IC das métricas
SHAP_SAMPLE_SIZE = 200       # amostra para acelerar SHAP (se muitos dados)
CLINICAL_COLUMN = None       # Nome da coluna com escore cognitivo (ex: 'MMSE'), se disponível

# ========== 1. REÚSO DA ESTRUTURA LOSO ==========
# Recriamos os grupos e dados
X_full = df_audit[FEATURE_COLS].copy().reset_index(drop=True)
y_full = (df_audit['label'] == 'AD').astype(int).values
groups = df_audit['subject_id'].values

# ========== 2. ESTRUTURAS PARA ARMAZENAR RESULTADOS ==========
results_per_fold = []          # lista de dicionários com preds, shap, etc.
modelos_por_fold = []          # guardar modelos treinados (opcional)

logo = LeaveOneGroupOut()
fold_idx = 0

# Para métricas agregadas
all_subject_auc = []
all_subject_sens = []
all_subject_spec = []
all_subject_bal_acc = []
all_subject_f1 = []

# Para SHAP e importância
shap_values_list = []          # lista de arrays (n_test_samples, n_features)
perm_importances_list = []     # lista de arrays (n_features,)
feature_names = FEATURE_COLS
n_features = len(feature_names)

# Para análise de erros
false_pos_subjects = []        # subject_id que foram FP
false_neg_subjects = []        # subject_id que foram FN

print("="*70)
print(" INICIANDO ANÁLISE LOSO COM EXPLICABILIDADE (SHAP + PERMUTATION)")
print("="*70)

# ========== 3. LOOP PRINCIPAL (CADA SUJEITO É TESTE UMA VEZ) ==========
for train_idx, test_idx in logo.split(X_full, y_full, groups):
    # Identifica o sujeito de teste
    test_subject = groups[test_idx[0]]   # todos os indices do mesmo sujeito
    print(f"\n▶ Fold {fold_idx+1}: Teste com sujeito {test_subject} "
          f"({len(test_idx)} épocas)")

    X_train = X_full.iloc[train_idx]
    X_test = X_full.iloc[test_idx]
    y_train = y_full[train_idx]
    y_test = y_full[test_idx]
    
    # Treina o Random Forest (substitua pelo seu modelo favorito)
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(n_estimators=200, max_depth=10,
                                random_state=RANDOM_STATE, n_jobs=-1)
    rf.fit(X_train, y_train)
    modelos_por_fold.append(rf)
    
    # --- Predições em nível de época ---
    probas_epoca = rf.predict_proba(X_test)[:, 1]
    preds_epoca = (probas_epoca >= 0.5).astype(int)
    
    # --- Agregação por sujeito (média das probabilidades) ---
    proba_subject = np.mean(probas_epoca)
    pred_subject = 1 if proba_subject >= 0.5 else 0
    y_true_subject = 1 if np.any(y_test == 1) else 0   # sujeito é AD se pelo menos uma época for AD
    
    # Métricas do sujeito
    auc_subj = roc_auc_score([y_true_subject], [proba_subject]) if len(np.unique(y_true_subject))>1 else np.nan
    sens_subj = recall_score([y_true_subject], [pred_subject], pos_label=1, zero_division=0)
    spec_subj = recall_score([y_true_subject], [pred_subject], pos_label=0, zero_division=0)
    bal_acc = balanced_accuracy_score([y_true_subject], [pred_subject])
    f1_subj = f1_score([y_true_subject], [pred_subject], zero_division=0)
    
    all_subject_auc.append(auc_subj)
    all_subject_sens.append(sens_subj)
    all_subject_spec.append(spec_subj)
    all_subject_bal_acc.append(bal_acc)
    all_subject_f1.append(f1_subj)
    
    # Armazena para análise de erros
    if pred_subject == 1 and y_true_subject == 0:
        false_pos_subjects.append(test_subject)
    elif pred_subject == 0 and y_true_subject == 1:
        false_neg_subjects.append(test_subject)
    
    # --- SHAP (apenas nos dados de teste) ---
    # Para acelerar, podemos subamostrar se X_test for grande
    if len(X_test) > SHAP_SAMPLE_SIZE:
        idx_sample = np.random.choice(len(X_test), SHAP_SAMPLE_SIZE, replace=False)
        X_test_sample = X_test.iloc[idx_sample]
    else:
        X_test_sample = X_test
    
    explainer = shap.TreeExplainer(rf)
    shap_test = explainer.shap_values(X_test_sample)[1]  # classe AD
    shap_values_list.append(shap_test)
    
    # --- Permutation Importance nos dados de teste ---
    perm_imp = permutation_importance(rf, X_test, y_test,
                                      n_repeats=N_REPEATS_PERM,
                                      scoring='roc_auc',
                                      random_state=RANDOM_STATE,
                                      n_jobs=-1)
    perm_importances_list.append(perm_imp.importances_mean)
    
    # Guarda predições para possível uso futuro
    results_per_fold.append({
        'fold': fold_idx,
        'test_subject': test_subject,
        'y_true': y_true_subject,
        'y_pred': pred_subject,
        'proba': proba_subject,
        'shap': shap_test,
        'perm_imp': perm_imp.importances_mean
    })
    
    fold_idx += 1

print("\n" + "="*70)
print(" FIM DO LOOP LOSO – AGREGANDO ESTATÍSTICAS")
print("="*70)

# ========== 4. MÉTRICAS DE DESEMPENHO AGREGADAS ==========
# Remove NaNs (caso algum fold tenha apenas uma classe no teste)
valid_mask = ~np.isnan(all_subject_auc)
metrics_df = pd.DataFrame({
    'AUC': np.array(all_subject_auc)[valid_mask],
    'Sensitivity': np.array(all_subject_sens)[valid_mask],
    'Specificity': np.array(all_subject_spec)[valid_mask],
    'Balanced_Accuracy': np.array(all_subject_bal_acc)[valid_mask],
    'F1': np.array(all_subject_f1)[valid_mask]
})

print("\n📊 DESEMPENHO MÉDIO (nível sujeito, LOSO):")
for col in metrics_df.columns:
    mean_val = metrics_df[col].mean()
    std_val = metrics_df[col].std()
    ci_low, ci_high = np.percentile(metrics_df[col], [2.5, 97.5])
    print(f"  {col:20s}: {mean_val:.4f} ± {std_val:.4f}  (95% CI: {ci_low:.4f}–{ci_high:.4f})")

# ========== 5. IMPORTÂNCIA GLOBAL DAS FEATURES ==========
# 5.1 SHAP (agregação de todos os folds)
shap_all = np.vstack(shap_values_list)  # (total_test_samples, n_features)
shap_global_mean = np.abs(shap_all).mean(axis=0)
shap_global_std = np.abs(shap_all).std(axis=0)
shap_direction = np.sign(shap_all).mean(axis=0)  # sinal médio

# 5.2 Permutation Importance
perm_all = np.array(perm_importances_list)  # (n_folds, n_features)
perm_mean = perm_all.mean(axis=0)
perm_std = perm_all.std(axis=0)

# 5.3 Consistência do sinal SHAP por feature (proporção de folds com mesmo sinal médio)
# Precisamos recalcular o sinal médio por fold para cada feature
shap_sign_consistency = []
for f_idx in range(n_features):
    signs_per_fold = []
    for fold_res in results_per_fold:
        shap_fold = fold_res['shap']  # (n_samples_fold, n_features)
        mean_shap_feat = np.mean(shap_fold[:, f_idx])
        signs_per_fold.append(np.sign(mean_shap_feat))
    # Proporção de folds com sinal positivo (ou negativo, depende da direção dominante)
    pos_prop = np.mean(np.array(signs_per_fold) == 1)
    neg_prop = np.mean(np.array(signs_per_fold) == -1)
    consistency = max(pos_prop, neg_prop)
    shap_sign_consistency.append(consistency)

# Cria DataFrame de importância
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'SHAP_abs_mean': shap_global_mean,
    'SHAP_abs_std': shap_global_std,
    'SHAP_direction': shap_direction,
    'SHAP_consistency': shap_sign_consistency,
    'Permutation_ΔAUC_mean': perm_mean,
    'Permutation_ΔAUC_std': perm_std
}).sort_values('SHAP_abs_mean', ascending=False)

print("\n🧠 TOP 10 FEATURES POR IMPORTÂNCIA (SHAP absoluto):")
print(importance_df.head(10).to_string(float_format="%.4f"))

# ========== 6. ESTABILIDADE ENTRE FOLDS (ICC) ==========
try:
    from pingouin import icc
    # Prepara dados para ICC: cada fold é um "avaliador", cada feature é um "item"
    icc_data = perm_all.T  # (n_features, n_folds)
    icc_val = icc(data=icc_data, targets=feature_names, raters=list(range(icc_data.shape[1])), kind='ICC2k')['ICC'][0]
    print(f"\n📈 Coeficiente de Correlação Intraclasse (ICC) da Permutation Importance: {icc_val:.3f}")
    if icc_val > 0.75:
        print("   → Excelente confiabilidade entre folds.")
    elif icc_val > 0.5:
        print("   → Confiabilidade moderada.")
    else:
        print("   → Baixa confiabilidade – as importâncias variam muito entre sujeitos.")
except:
    print("\n⚠️ Pingouin não instalado. Pule ICC. Instale com: pip install pingouin")

# ========== 7. ANÁLISE DE ERROS (FALSOS POSITIVOS E NEGATIVOS) ==========
print("\n🔍 ANÁLISE DE ERROS:")
print(f"  Falsos Positivos (HC → AD): {len(false_pos_subjects)} sujeitos: {false_pos_subjects}")
print(f"  Falsos Negativos (AD → HC): {len(false_neg_subjects)} sujeitos: {false_neg_subjects}")

if len(false_pos_subjects) > 0:
    # Recupera as features médias desses sujeitos
    fp_mask = df_audit['subject_id'].isin(false_pos_subjects)
    fp_features_mean = df_audit.loc[fp_mask, FEATURE_COLS].mean()
    print("\n  Perfil espectral médio dos falsos positivos (vs. HC verdadeiros):")
    hc_mask = (df_audit['label'] == 'HC') & (~df_audit['subject_id'].isin(false_pos_subjects))
    hc_features_mean = df_audit.loc[hc_mask, FEATURE_COLS].mean()
    diff = fp_features_mean - hc_features_mean
    top_diff = diff.abs().sort_values(ascending=False).head(5)
    for feat in top_diff.index:
        print(f"    {feat}: +{diff[feat]:.4f} (FP > HC)")

# ========== 8. CORRELAÇÃO COM DADOS CLÍNICOS (se disponível) ==========
if CLINICAL_COLUMN is not None and CLINICAL_COLUMN in df_audit.columns:
    # Agrega escore clínico por sujeito (média se houver múltiplas épocas)
    clinical_subj = df_audit.groupby('subject_id')[CLINICAL_COLUMN].mean()
    # Pega a probabilidade predita por sujeito (do último fold, ou podemos reconstruir)
    proba_by_subject = {}
    for res in results_per_fold:
        proba_by_subject[res['test_subject']] = res['proba']
    common_subjects = set(proba_by_subject.keys()) & set(clinical_subj.index)
    if len(common_subjects) > 0:
        probas = [proba_by_subject[s] for s in common_subjects]
        clinical_vals = [clinical_subj[s] for s in common_subjects]
        corr, p_val = stats.spearmanr(probas, clinical_vals)
        print(f"\n📉 Correlação de Spearman (probabilidade AD vs. {CLINICAL_COLUMN}): r = {corr:.3f}, p = {p_val:.4f}")
        if corr < -0.5 and p_val < 0.05:
            print("   → Forte correlação negativa: conforme piora o escore cognitivo, aumenta a chance de AD (consistente com neurobiologia).")
    else:
        print("\n⚠️ Não foi possível alinhar os dados clínicos com os sujeitos do LOSO.")
else:
    print("\n⚠️ Nenhuma coluna clínica definida (CLINICAL_COLUMN = None). Pule correlação.")

# ========== 9. GRÁFICOS ==========
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 9.1 Boxplot das métricas de desempenho entre folds
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
metrics_df.boxplot(ax=axes[0], rot=45)
axes[0].set_title("Distribuição das Métricas entre Sujeitos (LOSO)")
axes[0].set_ylabel("Valor")
# Adiciona pontos individuais
for i, col in enumerate(metrics_df.columns):
    y = metrics_df[col].dropna()
    x = np.random.normal(i+1, 0.04, size=len(y))
    axes[0].scatter(x, y, alpha=0.5, color='black', s=20)

# 9.2 Curvas ROC médias (opcional: plotar todas)
# Vamos reconstruir as probabilidades por sujeito para uma ROC global (não agregada)
# Mas cuidado: cada sujeito contribui com apenas um ponto. Para uma ROC suave, plotamos as curvas de cada fold?
# Melhor: mostrar a ROC de cada fold (linhas finas) e a média.
from sklearn.metrics import roc_curve
all_fpr = []
all_tpr = []
for res in results_per_fold:
    # Para cada fold, só temos um ponto (sujeito). Não dá curva. Portanto, pulamos ROC por fold.
    # Em vez disso, calculamos a curva usando as probabilidades das épocas (cuidado: correlação intra-sujeito)
    pass
# Como alternativa, plotamos a matriz de confusão agregada (nível sujeito)
y_true_all = [res['y_true'] for res in results_per_fold]
y_pred_all = [res['y_pred'] for res in results_per_fold]
cm = confusion_matrix(y_true_all, y_pred_all)
disp = ConfusionMatrixDisplay(cm, display_labels=['HC', 'AD'])
disp.plot(ax=axes[1], cmap='Blues', values_format='d')
axes[1].set_title("Matriz de Confusão (LOSO – nível sujeito)")
plt.tight_layout()
plt.show()

# 9.3 Importância das Features (SHAP e Permutation)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
# SHAP absoluto
top_n = min(15, n_features)
top_features = importance_df.head(top_n)
sns.barplot(data=top_features, x='SHAP_abs_mean', y='Feature', ax=axes[0], palette='viridis')
axes[0].errorbar(x=top_features['SHAP_abs_mean'], y=range(top_n), 
                 xerr=top_features['SHAP_abs_std'], fmt='none', color='black', capsize=3)
axes[0].set_title("Importância Global SHAP (|SHAP| médio ± DP entre amostras)")
axes[0].set_xlabel("|SHAP value| médio")
# Permutation importance
sns.barplot(data=top_features, x='Permutation_ΔAUC_mean', y='Feature', ax=axes[1], palette='rocket')
axes[1].errorbar(x=top_features['Permutation_ΔAUC_mean'], y=range(top_n),
                 xerr=top_features['Permutation_ΔAUC_std'], fmt='none', color='black', capsize=3)
axes[1].set_title("Permutation Importance (ΔAUC médio ± DP entre folds)")
axes[1].set_xlabel("Queda no AUC ao permutar a feature")
plt.tight_layout()
plt.show()

# 9.4 SHAP Summary Plot (global)
# Para não sobrecarregar, usamos uma amostra representativa
shap.summary_plot(shap_all, X_full[feature_names], show=False, max_display=15)
plt.title("SHAP Summary Plot (todos os testes agregados)")
plt.tight_layout()
plt.show()

# 9.5 Partial Dependence Plots para as 3 features mais importantes
top3_features = importance_df['Feature'].head(3).tolist()
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# Precisamos de um modelo final treinado com todos os dados para PDP? 
# Para não vazar, podemos treinar um modelo com todos os dados (apenas para visualização ilustrativa)
# Ou podemos calcular PDP dentro de cada fold e agregar. Para simplicidade, treinamos um modelo final.
# ATENÇÃO: Isso não é rigoroso para inferência causal, mas serve para visualizar tendência geral.
rf_final = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=RANDOM_STATE)
rf_final.fit(X_full, y_full)
for i, feat in enumerate(top3_features):
    PartialDependenceDisplay.from_estimator(rf_final, X_full, [feat], ax=axes[i], 
                                            grid_resolution=20, random_state=RANDOM_STATE)
    axes[i].set_title(f"PDP: {feat}")
plt.tight_layout()
plt.show()

# 9.6 Consistency plot: sinal consistente vs. importância
fig, ax = plt.subplots(figsize=(10, 6))
sc = ax.scatter(importance_df['SHAP_abs_mean'], importance_df['SHAP_consistency'], 
                c=importance_df['Permutation_ΔAUC_mean'], cmap='coolwarm', s=100, alpha=0.8)
for i, row in importance_df.iterrows():
    ax.annotate(row['Feature'], (row['SHAP_abs_mean'], row['SHAP_consistency']), 
                fontsize=8, alpha=0.7)
ax.set_xlabel("Importância SHAP absoluta")
ax.set_ylabel("Consistência do sinal (% folds com mesmo sinal)")
ax.set_title("Consistência Neurobiológica vs. Importância")
plt.colorbar(sc, label="Permutation ΔAUC")
plt.grid(True, alpha=0.3)
plt.show()

# 9.7 Perfil espectral dos erros (heatmap comparativo)
if len(false_pos_subjects) > 0 or len(false_neg_subjects) > 0:
    # Média das features para FP, FN, TP, TN
    groups_err = {}
    # Verdadeiros positivos (AD classificados como AD)
    tp_subj = [res['test_subject'] for res in results_per_fold if res['y_true']==1 and res['y_pred']==1]
    tn_subj = [res['test_subject'] for res in results_per_fold if res['y_true']==0 and res['y_pred']==0]
    
    for name, subj_list in [('FP (HC→AD)', false_pos_subjects),
                            ('FN (AD→HC)', false_neg_subjects),
                            ('TP (AD→AD)', tp_subj),
                            ('TN (HC→HC)', tn_subj)]:
        if subj_list:
            mask = df_audit['subject_id'].isin(subj_list)
            groups_err[name] = df_audit.loc[mask, FEATURE_COLS].mean()
        else:
            groups_err[name] = pd.Series(0, index=FEATURE_COLS)
    
    err_df = pd.DataFrame(groups_err).T
    # Normalizar por feature para visualização
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    err_df_scaled = pd.DataFrame(scaler.fit_transform(err_df.T).T, 
                                 index=err_df.index, columns=err_df.columns)
    plt.figure(figsize=(12, 6))
    sns.heatmap(err_df_scaled, cmap='RdBu_r', center=0, annot=False, cbar_kws={'label': 'Z-score'})
    plt.title("Perfil Espectral dos Erros (média por grupo, padronizado por feature)")
    plt.ylabel("Grupo")
    plt.xlabel("Features")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

print("\n✅ Análise completa finalizada!")
```
