**DESEMPENHO NO MOCA (HC x AD)**

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Pasta principal onde vamos procurar
pasta_procura = "C:/Users/tiago/Downloads"

# Nomes dos arquivos de cognição
nome_cog_hc = "cognition_hc_eeg_data.csv"
nome_cog_ad = "cognition_ad_eeg_data.csv"

caminho_cog_hc = None
caminho_cog_ad = None

print("Procurando os arquivos de testes cognitivos...\n")

# Vasculha a pasta Downloads atrás dos dois arquivos
for raiz, diretorios, arquivos in os.walk(pasta_procura):
    if nome_cog_hc in arquivos and caminho_cog_hc is None:
        caminho_cog_hc = os.path.join(raiz, nome_cog_hc).replace("\\", "/")
    if nome_cog_ad in arquivos and caminho_cog_ad is None:
        caminho_cog_ad = os.path.join(raiz, nome_cog_ad).replace("\\", "/")

# Se encontrou os dois arquivos, vamos à análise!
if caminho_cog_hc and caminho_cog_ad:
    # 1. Carrega os dados
    df_cog_hc = pd.read_csv(caminho_cog_hc)
    df_cog_hc['Grupo'] = 'Saudáveis (HC)' # Cria uma etiqueta para eles
    
    df_cog_ad = pd.read_csv(caminho_cog_ad)
    df_cog_ad['Grupo'] = 'Alzheimer (AD)' # Cria uma etiqueta para eles
    
    # 2. Junta tudo numa tabela só
    df_completo = pd.concat([df_cog_hc, df_cog_ad], ignore_index=True)
    
    # 3. Descobre automaticamente quais colunas são números (os testes)
    # Ignorando colunas de texto ou o ID do sujeito
    colunas_numericas = df_completo.select_dtypes(include=['float64', 'int64']).columns.tolist()
    colunas_testes = [col for col in colunas_numericas if 'id' not in col.lower() and 'unnamed' not in col.lower()]
    
    print("✅ Dados carregados e unidos com sucesso!")
    print(f"🧠 Testes identificados na tabela: {colunas_testes}\n")
    
    # 4. Plota os Boxplots para até os 4 primeiros testes encontrados
    if len(colunas_testes) > 0:
        num_plots = min(4, len(colunas_testes))
        fig, axes = plt.subplots(1, num_plots, figsize=(15, 6))
        
        # Garante que 'axes' seja uma lista mesmo se houver só 1 teste
        if num_plots == 1: axes = [axes]
            
        for i in range(num_plots):
            teste = colunas_testes[i]
            # O Pandas tem uma função direta para criar boxplots
            df_completo.boxplot(column=teste, by='Grupo', ax=axes[i], grid=False, 
                                boxprops=dict(linewidth=2, color='blue'),
                                medianprops=dict(linewidth=2, color='red'))
            
            axes[i].set_title(teste, fontsize=12)
            axes[i].set_xlabel('')
            axes[i].set_ylabel('Pontuação', fontsize=10)
        
        # Ajustes visuais
        plt.suptitle('Desempenho Cognitivo: Saudáveis vs Alzheimer', fontsize=16, y=1.05)
        plt.tight_layout()
        plt.show()
    else:
        print("❌ Não encontrei colunas numéricas com resultados de testes.")

else:
    print("❌ Faltou algum arquivo de cognição.")
    if not caminho_cog_hc: print(f"Não encontrei o: {nome_cog_hc}")
    if not caminho_cog_ad: print(f"Não encontrei o: {nome_cog_ad}")
```

**TESTES ESTATISTICOS**

**Célula 1: Diferença Estatística (Teste T e Correlação Ponto-Bisserial)**
Objetivo: Provar matematicamente se a nota dos pacientes com Alzheimer é significativamente diferente da nota dos saudáveis.

```python
import os
import pandas as pd
from scipy import stats

# 1. Busca Automática para garantir que os arquivos sejam encontrados
pasta_procura = "C:/Users/tiago/Downloads"
arquivos_alvo = ["cognition_hc_eeg_data.csv", "cognition_ad_eeg_data.csv"]
caminhos = {}

for raiz, dirs, files in os.walk(pasta_procura):
    for nome in arquivos_alvo:
        if nome in files and nome not in caminhos:
            caminhos[nome] = os.path.join(raiz, nome).replace("\\", "/")

if len(caminhos) == 2:
    # 2. União dos dados
    df_hc = pd.read_csv(caminhos["cognition_hc_eeg_data.csv"])
    df_ad = pd.read_csv(caminhos["cognition_ad_eeg_data.csv"])
    df_hc['Grupo'] = 'Saudável (HC)'
    df_ad['Grupo'] = 'Alzheimer (AD)'
    df_cog = pd.concat([df_hc, df_ad], ignore_index=True)
    df_cog['Diagnostico_Binario'] = df_cog['Grupo'].apply(lambda x: 1 if 'Alzheimer' in x else 0)
    
    # 3. Cálculo de Estatísticas para cada teste
    colunas_testes = [col for col in df_cog.select_dtypes(include=['number']).columns 
                      if 'id' not in col.lower() and 'binario' not in col.lower()]
    
    resultados = []
    for teste in colunas_testes:
        df_limpo = df_cog.dropna(subset=[teste])
        # Correlação Ponto-Bisserial
        corr, _ = stats.pointbiserialr(df_limpo['Diagnostico_Binario'], df_limpo[teste])
        # Teste T para diferença de médias
        g_hc = df_limpo[df_limpo['Diagnostico_Binario'] == 0][teste]
        g_ad = df_limpo[df_limpo['Diagnostico_Binario'] == 1][teste]
        t_stat, p_val = stats.ttest_ind(g_hc, g_ad, equal_var=False)
        
        resultados.append({
            'Teste': teste,
            'Média Saudáveis': round(g_hc.mean(), 2),
            'Média Alzheimer': round(g_ad.mean(), 2),
            'Correção': round(corr, 3),
            'Valor-p (Significância)': round(p_val, 5),
            'Diferença Real?': 'Sim ✅' if p_val < 0.05 else 'Não ❌'
        })
    
    df_resumo = pd.DataFrame(resultados).sort_values(by='Valor-p (Significância)')
    print("TABELA 1: RESUMO ESTATÍSTICO (COGNITIVO vs DIAGNÓSTICO)")
    display(df_resumo)
else:
    print("Erro: Não encontrei os arquivos 'cognition_hc_eeg_data.csv' e 'cognition_ad_eeg_data.csv' nos Downloads.")
```

**Célula 2: Curva ROC**
Objetivo: Este gráfico mostra a "sensibilidade" do teste. Quanto mais a linha vermelha "sobe" para o canto esquerdo, mais poderoso é o teste para detectar Alzheimer sem errar.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Usaremos o teste mais significativo da tabela anterior
melhor_teste = df_resumo.iloc[0]['Teste']

df_plot = df_cog.dropna(subset=[melhor_teste])
X = df_plot[[melhor_teste]]
y = df_plot['Diagnostico_Binario']

# Criando o modelo de previsão
modelo = LogisticRegression().fit(X, y)
probs = modelo.predict_proba(X)[:, 1]

fpr, tpr, _ = roc_curve(y, probs)
auc = roc_auc_score(y, probs)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='red', lw=3, label=f'Poder de Detecção (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.title(f'GRÁFICO 1: Capacidade do teste {melhor_teste.upper()} em diagnosticar AD')
plt.xlabel('Falso Positivo (Classificar Saudável como AD)')
plt.ylabel('Verdadeiro Positivo (Classificar AD corretamente)')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.show()

print(f"Interpretação: O teste {melhor_teste} tem {auc*100:.1f}% de chance de separar corretamente um paciente AD de um Saudável.")
```
